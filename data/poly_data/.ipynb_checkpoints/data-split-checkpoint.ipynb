```python
# Cell 1: inline splitter + run it
import json, random
from pathlib import Path
from typing import List, Dict, Any

def split_dataset(
    records: List[Dict[str, Any]],
    force_train_id: str = "mp-5229",
    seed: int = 42,
    train_frac: float = 0.8,
    val_frac: float = 0.1,
):
    n = len(records)
    if n == 0:
        return [], [], []

    forced = [r for r in records if r.get("id") == force_train_id]
    pool = [r for r in records if r.get("id") != force_train_id]

    random.seed(seed)
    random.shuffle(pool)

    n_train_target = int(round(train_frac * n))
    n_val_target = int(round(val_frac * n))
    # n_test_target = n - n_train_target - n_val_target

    n_train_from_pool = max(0, n_train_target - len(forced))
    n_train_from_pool = min(n_train_from_pool, len(pool))

    train = forced + pool[:n_train_from_pool]
    start = n_train_from_pool
    end = start + min(n_val_target, max(0, len(pool) - start))
    val = pool[start:end]
    test = pool[end:]

    # final sanity
    assert len(train) + len(val) + len(test) == n, "Split sizes do not sum to total!"
    return train, val, test

# CONFIG: adjust input path and output directory here
input_path = Path("./dataset_8_motifs_single.json")   # <- change to your file
outdir = Path("./")                            # <- change to desired folder
seed = 42
force_train_id = "mp-5229"

# ensure outdir exists
outdir.mkdir(parents=True, exist_ok=True)

# load data
with input_path.open("r") as f:
    data = json.load(f)
assert isinstance(data, list), "input must be a top-level JSON list"

# split
train, val, test = split_dataset(
    data,
    force_train_id=force_train_id,
    seed=seed,
)

# write outputs
def write_list(lst, fname):
    p = outdir / fname
    with p.open("w") as f:
        json.dump(lst, f, indent=2)
    print(f"Wrote {len(lst)} entries to {p}")

write_list(train, "train_polyhedra.json")
write_list(val, "val_polyhedra.json")
write_list(test, "test_polyhedra.json")
```

```python
# Cell: simple deterministic split (forced entry first in train)
import json, random
from pathlib import Path
from typing import List, Dict, Any

# CONFIG
input_path = Path("./dataset_8_motifs_single.json")  # change if needed
outdir = Path("./")                                  # change if needed
seed = 42
force_train_id = "mp-5229"
train_target = 1000
val_target = 300
test_target = 300

outdir.mkdir(parents=True, exist_ok=True)

# load data
with input_path.open("r") as f:
    data: List[Dict[str, Any]] = json.load(f)

# find forced entry (assumed present)
forced = next(r for r in data if r.get("id") == force_train_id)

# build pool excluding forced
pool = [r for r in data if r.get("id") != force_train_id]

# deterministic shuffle
random.seed(seed)
random.shuffle(pool)

# select splits (no extra logic; dataset is assumed large enough)
train = [forced] + pool[: (train_target - 1)]
val = pool[(train_target - 1) : (train_target - 1 + val_target)]
test = pool[(train_target - 1 + val_target) : (train_target - 1 + val_target + test_target)]

# write outputs
def write_json(lst, fname):
    p = outdir / fname
    with p.open("w") as f:
        json.dump(lst, f, indent=2)
    print(f"Wrote {len(lst)} entries to {p}")

write_json(train, "train_polyhedra_small.json")
write_json(val, "val_polyhedra_small.json")
write_json(test, "test_polyhedra_small.json")

print(f"Final counts -> train: {len(train)}, val: {len(val)}, test: {len(test)}")
```

```python
import json
import re
import os
from glob import glob

# Config
base_dir = "../../../Diffhedron/motifs_alex_mp_20/"  # e.g. "/mnt/data/json_datasets"
search_pattern = "**/*_single.json"  # match all files ending with _single.json
output_fname = "one_polyhedron_data.json"

# === MAIN SCRIPT ===
if not os.path.isdir(base_dir):
    raise NotADirectoryError(f"The specified folder does not exist: {base_dir}")

# Build the search pattern
search_path = os.path.join(base_dir, search_pattern)

# Recursively find all matching JSON files
files = glob(search_path, recursive=True)
if not files:
    raise FileNotFoundError(f"No files matching pattern '{search_pattern}' were found in '{base_dir}'.")

seen_base = set()
out_samples = []
files_scanned = 0
errors = []
files.sort()

for filepath in files:
    print(filepath)
    files_scanned += 1
    try:
        with open(filepath, "r") as fh:
            data = json.load(fh)
    except Exception as e:
        errors.append((filepath, str(e)))
        print(f"Warning: failed to read/parse '{filepath}': {e}")
        continue

    # Expecting data to be a list of crystals (like your examples)
    if not isinstance(data, list):
        print(f"Warning: file '{filepath}' does not contain a JSON list at top-level; skipping.")
        continue

    for crystal in data:
        id = crystal.get("id", "")
        composition = crystal.get("composition", "")
        for poly in crystal.get("polyhedra", []):
            label = poly.get("label", "")
            base_label = re.sub(r'_[0-9]+$', '', label)
            if base_label in seen_base:
                continue
            seen_base.add(base_label)
            # create a new sample (keep the original poly object unchanged, preserving suffix)
            sample = {
                "id": id,
                "composition": composition,
                "polyhedra": [poly]
            }
            out_samples.append(sample)

# Save output JSON
with open(output_fname, "w") as outfh:
    json.dump(out_samples, outfh, indent=2)

# Summary
print("=== Summary ===")
print(f"Files scanned (matching pattern): {len(files)}")
print(f"Files successfully opened: {files_scanned - len(errors)}")
print(f"Files with errors: {len(errors)}")
if errors:
    for fp, err in errors:
        print(f" - {fp}: {err}")
print(f"Unique base polyhedron types retained: {len(out_samples)}")
print(f"Output saved to: {os.path.abspath(output_fname)}")
```

```python
import json
import re
import os
from glob import glob

# Config
base_dir = "../../../Diffhedron/motifs_alex_mp_20/"  # e.g. "/mnt/data/json_datasets"
search_pattern = "**/*_single.json"  # match all files ending with _single.json
output_fname = "one_polyhedron_data.json"

# === MAIN SCRIPT ===
if not os.path.isdir(base_dir):
    raise NotADirectoryError(f"The specified folder does not exist: {base_dir}")

# Build the search pattern
search_path = os.path.join(base_dir, search_pattern)

# Recursively find all matching JSON files
files = glob(search_path, recursive=True)
if not files:
    raise FileNotFoundError(f"No files matching pattern '{search_pattern}' were found in '{base_dir}'.")

# Dictionary to store statistics and samples
# Structure:
# {
#   "BaseLabelString": {
#       "MotifSymbolString": { 
#           "count": int, 
#           "sample_poly": dict, 
#           "parent_info": {"id": str, "composition": str} 
#       }
#   }
# }
poly_groups = {}

files_scanned = 0
errors = []
files.sort()

print("Scanning files and aggregating motif statistics...")

for filepath in files:
    print(f"Scanning: {filepath}")
    
    files_scanned += 1
    try:
        with open(filepath, "r") as fh:
            data = json.load(fh)
    except Exception as e:
        errors.append((filepath, str(e)))
        print(f"Warning: failed to read/parse '{filepath}': {e}")
        continue

    # Expecting data to be a list of crystals
    if not isinstance(data, list):
        print(f"Warning: file '{filepath}' does not contain a JSON list at top-level; skipping.")
        continue

    for crystal in data:
        c_id = crystal.get("id", "")
        composition = crystal.get("composition", "")
        
        for poly in crystal.get("polyhedra", []):
            label = poly.get("label", "")
            motif_symbol = poly.get("motif_symbol", "unknown")
            
            # Extract base label (e.g., "Tb1P6_0" -> "Tb1P6")
            base_label = re.sub(r'_[0-9]+$', '', label)
            
            # Initialize dictionary structure if new
            if base_label not in poly_groups:
                poly_groups[base_label] = {}
            
            if motif_symbol not in poly_groups[base_label]:
                poly_groups[base_label][motif_symbol] = {
                    "count": 0,
                    "sample_poly": poly, # Keep the first one of this specific motif type we see
                    "parent_info": {
                        "id": c_id, 
                        "composition": composition
                    }
                }
            
            # Increment the counter for this specific shape
            poly_groups[base_label][motif_symbol]["count"] += 1

# === SELECTION PHASE ===
print("Selecting most frequent motifs...")

out_samples = []

for base_label, motifs_data in poly_groups.items():
    # Find the motif_symbol with the highest count for this base_label
    # motifs_data is a dict where keys are motif_symbols (e.g., "O:6", "T:4")
    best_motif = max(motifs_data, key=lambda k: motifs_data[k]["count"])
    
    # Retrieve the stored data for the winner
    winner_data = motifs_data[best_motif]
    
    # Create the sample object
    sample = {
        "id": winner_data["parent_info"]["id"],
        "composition": winner_data["parent_info"]["composition"],
        "polyhedra": [winner_data["sample_poly"]]
    }
    
    out_samples.append(sample)

# Save output JSON
with open(output_fname, "w") as outfh:
    json.dump(out_samples, outfh, indent=2)

# Summary
print("=== Summary ===")
print(f"Files scanned: {files_scanned}")
print(f"Files with errors: {len(errors)}")
if errors:
    for fp, err in errors:
        print(f" - {fp}: {err}")
print(f"Unique base polyhedron types processed: {len(out_samples)}")
print(f"Output saved to: {os.path.abspath(output_fname)}")
```

```python
import json
import os
from random import Random

# === USER CONFIG ===
input_fname = "one_polyhedron_data.json"   # existing file (list of samples)
output_train = "train_one_polyhedron.json"
output_test  = "test_one_polyhedron.json"
output_val   = "val_one_polyhedron.json"

RSEED = 42  # for reproducibility
rnd = Random(RSEED)

# === load input ===
if not os.path.exists(input_fname):
    raise FileNotFoundError(f"Input file not found: {input_fname}")

with open(input_fname, "r") as fh:
    data = json.load(fh)

if not isinstance(data, list):
    raise RuntimeError(f"Expected top-level list in {input_fname}, got {type(data)}")

total = len(data)
if total < 2:
    raise RuntimeError("Input must contain at least 2 samples (so we can skip the first two).")

# === take aside the first two samples (they will be forced into train) ===
manual_train_samples = data[:2]
remaining = data[2:]   # list of samples to be split randomly

n_rem = len(remaining)
if n_rem == 0:
    # nothing to split — just place the two into train and write empty test/val
    train_samples = list(manual_train_samples)
    test_samples = []
    val_samples = []
else:
    # shuffle remaining samples reproducibly
    rnd.shuffle(remaining)

    # Compute counts:
    # Interpret "80-20-20" as: 80% train, remaining 20% split equally into test and val
    desired_train_count = int(round(0.80 * total))
    # ensure at least manual samples are included in train later
    # Train should include the manual two plus some from remaining to reach desired_train_count
    # Number of samples to take from remaining into train:
    need_from_remaining_for_train = max(0, desired_train_count - len(manual_train_samples))
    # Cap it by n_rem
    take_train = min(need_from_remaining_for_train, n_rem)

    train_from_remaining = remaining[:take_train]
    leftover_after_train = remaining[take_train:]

    # Split leftover into test and val equally
    n_left = len(leftover_after_train)
    test_count = n_left // 2
    val_count = n_left - test_count  # remaining goes to val (handles odd counts)

    test_samples = leftover_after_train[:test_count]
    val_samples = leftover_after_train[test_count:test_count + val_count]

    # Build final train: manual two + selected train_from_remaining
    train_samples = list(manual_train_samples) + train_from_remaining

# Optionally: if you prefer to ensure train contains at least 80% of TOTAL even when rounding issues occur,
# you can move samples from val/test into train. Current code aims to meet desired_train_count via rounding.

# === save outputs (in current working dir) ===
with open(output_train, "w") as fh:
    json.dump(train_samples, fh, indent=2)
with open(output_test, "w") as fh:
    json.dump(test_samples, fh, indent=2)
with open(output_val, "w") as fh:
    json.dump(val_samples, fh, indent=2)

# === summary ===
print("=== Split summary ===")
print(f"Input file: {os.path.abspath(input_fname)}")
print(f"Total samples in input: {total}")
print(f"Manual (first 2) samples forced into train: {len(manual_train_samples)}")
print(f"Remaining samples to split: {n_rem}")
print(f"Final counts -> train: {len(train_samples)}, test: {len(test_samples)}, val: {len(val_samples)}")
print(f"Saved -> {os.path.abspath(output_train)}")
print(f"Saved -> {os.path.abspath(output_test)}")
print(f"Saved -> {os.path.abspath(output_val)}")
```
